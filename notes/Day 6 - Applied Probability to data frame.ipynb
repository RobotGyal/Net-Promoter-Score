```python
def conditional_prob(value1, value2, condition):
    set1 = df_tennis[df_tennis[value1] == condition][value2]
    set1 = set1.unique()
    pop = len(set1)
    prob={}
    for i in set1:
        prob[i] = len(df_tennis[(df_tennis[value1] == condition) & (df_tennis[value2]== i)]) / (pop)
        print(prob[i])
    return prob
        
        
print(conditional_prob('Wind', 'Decision', 'Weak'))
    
```
